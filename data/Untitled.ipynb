```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
training_data_path = "train.csv"
testing_data_path = "test.csv"
```

```python
# Get headers of the CSV training data
with open(training_data_path, "r") as f:
    headers = [h.strip() for h in f.readline().split(',')[2:]]

# import CSV training data skipping the 2 first columns (id and prediction)
features = np.genfromtxt(training_data_path,delimiter=",", skip_header=1)[:, 2:]
labels = np.genfromtxt(training_data_path, dtype=str, delimiter=",", usecols = (1),skip_header=1)
```

```python
def show_histogram(features, labels, title):
    
    mini = np.min(features)
    maxi = np.max(features)
    
    background = features[labels == 'b']
    signal = features[labels == 's']
    
    info = """
    =================== {} ===================
    Overall:      mean {:4.2f}, median {:4.2f}, variance {:4.2f}
    Background: mean {:4.2f}, median {:4.2f}, variance {:4.2f}
    Signal:       mean {:4.2f}, median {:4.2f}, variance {:4.2f}
    ==========================================
    """.format(
        title,
        np.mean(features), np.median(features), np.var(features),
        np.mean(background), np.median(background), np.var(background),
        np.mean(signal), np.median(signal), np.var(signal)
    )
    
    print(info)
    
    
    plt.hist([features, background, signal], label=["All", "Background", "Signal"], stacked=True)
    plt.legend()
    plt.title(title)
```

```python
for i, h in enumerate(headers):
    a = plt.subplotd(3,10, i+1)
    show_histogram(features[:,i], labels, h)
```
